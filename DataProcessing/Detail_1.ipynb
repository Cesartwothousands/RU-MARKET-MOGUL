```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# 
end_date = datetime.now()+timedelta(days=1)
start_date = end_date - timedelta(days=30)

# AAPL
aapl = yf.Ticker('AAPL')

# AAPL
aapl_history = aapl.history(start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))
print(aapl_history)
# 
plt.figure(figsize=(14, 7))
plt.plot(aapl_history['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('AAPL Stock Price in the Last Month')
plt.grid()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# AAPL
aapl = yf.Ticker("AAPL")

# 
day_prices = aapl.history(period="1d")

# 
five_day_prices = aapl.history(period="5d")

# 
month_prices = aapl.history(period="1mo")

# 
year_prices = aapl.history(period="1y")

# 
five_year_prices = aapl.history(period="5y")

# 
plt.plot(day_prices['Close'])
plt.title("AAPL Daily Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

# 
plt.plot(five_day_prices['Close'])
plt.title("AAPL 5-Day Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

# 
plt.plot(month_prices['Close'])
plt.title("AAPL Monthly Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

# 
plt.plot(year_prices['Close'])
plt.title("AAPL Yearly Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()

# 
plt.plot(five_year_prices['Close'])
plt.title("AAPL 5-Year Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```
