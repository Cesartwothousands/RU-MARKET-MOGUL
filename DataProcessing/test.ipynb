```python
import yfinance as yf
import json
import pandas as pd

# Define the stock symbols you want to fetch data for
symbols = pd.read_csv('Top100_2.txt', header=None)
symbols = symbols[0].tolist()

# Parse the data and format it as per your column definitions
data = []

for symbol in symbols:
    stock_info = yf.Ticker(symbol).info
    
    last_price = round(stock_info['currentPrice'], 2)
    previous_close = round(stock_info['previousClose'], 2)
    
    # Calculate price change in absolute and percentage terms
    price_change = round(last_price - previous_close, 2)
    price_change_percent = round((price_change / previous_close) * 100, 2) if previous_close else 0
    
    volume = round(stock_info['regularMarketVolume'], 2)
        
    marketcap = round(stock_info['marketCap'], 2)
    
    industry = stock_info['industry']

    data.append({
        'symbol': symbol,
        'name': stock_info['longName'],
        'sector': industry,
        'lastprice': last_price,
        'change1': price_change,
        'change2': price_change_percent,
        'volume': volume,
        'marketcap': marketcap
    })

# Convert the data to JSON
json_data = json.dumps(data)
print(json_data)
```
